```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload gambar
from google.colab import files
uploaded = files.upload()

# Baca gambar grayscale
filename = list(uploaded.keys())[0]
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# --- Transformasi ---
negatif = 255 - img

c = 255 / np.log(1 + np.max(img))
log_img = c * np.log(1 + img)
log_img = np.array(log_img, dtype=np.uint8)

# --- Tampilkan gambar ---
plt.figure(figsize=(12,10))

plt.subplot(3,3,1)
plt.title("Citra Asli")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(3,3,2)
plt.title("Negatif")
plt.imshow(negatif, cmap='gray')
plt.axis('off')

plt.subplot(3,3,3)
plt.title("Logaritmik")
plt.imshow(log_img, cmap='gray')
plt.axis('off')

# --- Histogram masing-masing ---
plt.subplot(3,3,4)
plt.title("Histogram Asli")
plt.hist(img.ravel(), bins=256, range=[0,256])

plt.subplot(3,3,5)
plt.title("Histogram Negatif")
plt.hist(negatif.ravel(), bins=256, range=[0,256])

plt.subplot(3,3,6)
plt.title("Histogram Logaritmik")
plt.hist(log_img.ravel(), bins=256, range=[0,256])

plt.tight_layout()
plt.show()
```
